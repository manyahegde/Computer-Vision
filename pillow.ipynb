```python
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Open an image
image = Image.open("image.webp")
```

```python
image
```

```python
image.format
```

```python
# Resize the image to a specific size
resized_image = image.resize((300,200))
```

```python
resized_image
```

```python
# Convert the image to grayscale
grayscale_image = image.convert("L")
```

```python
grayscale_image
```

```python
# Define the cropping coordinates (left, upper, right, lower)
box = (100, 100, 400, 300)

# Crop the image
cropped_image = image.crop(box)
```

```python
cropped_image
```

```python
# Apply a blur filter
from PIL import ImageFilter
blurred_image = image.filter(ImageFilter.BLUR)
```

```python
blurred_image
```

```python
from PIL import ImageDraw, ImageFont
from IPython.display import display

# Create an ImageDraw object
draw = ImageDraw.Draw(image)

# Define the text and font
text = "Astral Projection!"
font = ImageFont.truetype("arial.ttf", size=40)

# Specify the position to add the text
x, y = 50, 50

# Add the text to the image with a different text color (e.g., blue)
draw.text((x, y), text, fill="white", font=font)

# Display the image directly in the Jupyter Notebook
display(image)
```

```python
# Save the image with the added text
image.save("image.jpg")
```
